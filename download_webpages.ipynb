```python
import requests
```

```python
# Webpage that will be downloaded
url = "https://en.wikipedia.org/wiki/Leica_M7"

response = requests.get(url)
if response.status_code == 200:
    # Save the webpage as HTML file
    with open("/home/txie/CS600/project/webpages/leica_m7.html", "w", encoding="utf-8") as file:
        file.write(response.text)
    print("Webpage saved.")
else:
    print(f"Failed: {response.status_code}")
```
